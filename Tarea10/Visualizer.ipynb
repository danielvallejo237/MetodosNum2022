```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
file1="./problema3/parejas.txt"
with open(file1) as f:
    f1=f.readlines()
```

```python
puntos=[]
for p in f1[:10000]:
    a,b=p.split(',')
    puntos.append((float(a),float(b)))
puntos=np.array(puntos).transpose()
```

```python
def f(x):
    return x**3+42*x

pts=[f(p) for p in np.linspace(-3,2,num=1000)]
```

```python
ind=[]
for i in range(puntos.shape[1]):
    if 0<=puntos[1][i] and puntos[1][i]<=f(puntos[0][i]):
        ind.append("#7BBB12")
    elif 0>puntos[1][i] and puntos[1][i]>=f(puntos[0][i]):
        ind.append("#7BBB12")
    else:
        ind.append("magenta")
```

```python
plt.figure(figsize=(8,8))
plt.plot(np.linspace(-3,2,1000),pts,color='black',linewidth=4,label='groundTruth')
plt.scatter(puntos[0],puntos[1],color=ind,alpha=1.0,label='Puntos generados')
plt.legend()
plt.title("Integral Simple con Método de Monte Carlo",fontsize=20)
plt.savefig("Integracion1dMonteCarlo.jpg")
```

```python
with open("./problema4/pairs.txt") as p:
    parejas=p.readlines()
```

```python
parejas=parejas[:100000]
fPar=[]
for p in parejas:
    a,b,c=p.split(',')
    fPar.append((float(a),float(b),float(c)))
fPar=np.array(fPar).transpose()
```

```python
fPar.shape
```

```python
def g(x,y):
    return x**2+y**2

def indic(x,y,z):
    return (z<0 and z>=g(x,y)) or (z>=0 and z<=g(x,y))
```

```python
import math
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
x = np.arange(-3, 3, 0.2)
y = np.arange(-3, 3, 0.2)
X, Y = np.meshgrid(x, y)
Z =X**2+Y**2

ind=[]
for i in range(fPar[0].shape[0]):
    ib=indic(fPar[0][i],fPar[1][i],fPar[2][i])
    if(ib):
        ind.append("magenta")
    else:
        ind.append("#7BBB12")
ax.scatter(fPar[0],fPar[1],fPar[2],alpha=0.07,color=ind)
surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)
plt.title("Integral doble calculada con el método de Monte Carlo",fontsize=20)
plt.savefig("MonteCarlo2d.jpg")
```

# Problema 2

```python
with open("./problema2/ipol.txt") as f:
    file=f.readlines()
```

```python
def f(x):
    return x+ (x*math.sin(x/2.0))/3
```

```python
x=[]
y=[]
for fi in file[:-1]:
    a,b=fi.split(',')
    x.append(float(a))
    y.append(float(b))

y_real=[f(xi) for xi in x]
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y,color='blue',linewidth=3,label='FEM Int lambda=0.9')
plt.plot(x,y_real,color='red',linewidth=3,label='Ground Truth')
plt.title("Interpolación con Elemento Finito",fontsize=20)
plt.grid((50,50),color='black')
plt.legend()
plt.savefig("intfem.jpg")
```
