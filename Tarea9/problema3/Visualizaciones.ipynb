```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
with open ("../problema2/quadpuntos.txt","r") as f:
    puntos=f.readlines()
with open("../problema2/interpolpoints.txt","r") as k:
    inter=k.readlines()
```

```python
inx=[]
iny=[]
for p in inter:
    a,b=p.split(',')
    inx.append(float(a))
    iny.append(float(b))
```

```python
xs=[]
ys=[]
for p in puntos[:-1]:
    a,b=p.split(',')
    xs.append(float(a))
    ys.append(float(b))
```

```python
import math
def f(x):
    return 3.45*math.sin(x)+2.5*math.sqrt(np.exp(x))-3*x
def g(x):
    return x+ (x*math.sin(x/2.0))/3;
```

```python
gt=[g(p) for p in xs]
```

```python
plt.figure(figsize=(10,10))
plt.plot(xs,gt,color='blue',label='Ground Truth',linewidth=3.25,alpha=0.7)
plt.plot(xs,ys,color='red',alpha=0.8,label="Spline Cuadratico",linewidth=3.5)
plt.scatter(inx,iny,color='green',linewidth=5,label="Punto de interpolacion")
plt.grid((50,50),color='black')
plt.legend()
plt.title("Splines Cuadratico")
plt.savefig("splinescuadg.jpg")
```
