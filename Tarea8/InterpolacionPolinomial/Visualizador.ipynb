```python
import numpy as np
import os
import matplotlib.pyplot as plt
import math
```

```python
with open("Lagrange_1000_puntos_de_10_grados.txt","r") as f:
    salida=f.readlines()
```

```python
def f(x):
    return 3.45*math.sin(x)+2.5*math.sqrt(math.exp(x))-3*x
```

```python
salida
```

```python
salida=salida[1:]
xs=[]
ys=[]
ysreal=[]
for x in salida:
    a,b=x.split()
    xs.append(float(a))
    ys.append(float(b))
    ysreal.append(f(float(a)))
```

```python
plt.plot(xs,ys,color='blue',label='Interpolado',linewidth=4)
plt.plot(xs,ysreal,color='red',label='Ground Truth')
plt.legend()
plt.title("Interpolacin Lagrange 10 grados")
plt.savefig("LAGW10D.jpg")
```
