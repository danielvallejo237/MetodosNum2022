```python
import matplotlib.pyplot as plt
import sys
import numpy as np
```

```python
import math
def function1(x):
    return 14*x+4
def function2(x):
    return 3.23*(x*x*x)-0.25*(x*x)-10*x +12.47
def function3(x):
    return 3.45*math.sin(x)+2.5*math.sqrt(np.exp(x))-3*x
```

```python
real=list(np.linspace(-1,40,1500))
realF=[function1(x) for x in real]
```

```python
with open("SalidaInterpol_10_puntos_10_grados.txt","r") as f:
    file=f.readlines()
file=file[1:]
```

```python
x=[]
y=[]
for p in file:
    a,b=p.split()
    x.append(float(a))
    y.append(float(b))
```

```python
plt.figure(figsize=(10,10))
plt.plot(real,realF,color='blue',linewidth=2.0,label='real')
plt.scatter(x,y,color='red',linewidths=2.5,alpha=0.6,label='interpolado')
plt.title("Interpolacion de polinomio de grado 1")
plt.legend()
plt.savefig("Interpolacion1deg.jpg")
```
