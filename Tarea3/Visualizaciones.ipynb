```python
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
```

```python
with open("salida_der.txt") as f:
    sal=f.read()
    sal=[float(s) for s in sal.split()]
with open("salida_2der.txt") as f:
    sal2=f.read()
    sal2=[float(s) for s in sal2.split()]
sal3=[x**4+1 for x in list(np.linspace(-10,10,1000))]
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.linspace(-10,10,1000),sal,color='blue',label='Primera derivada estimada',linewidth=4)
plt.plot(np.linspace(-10,10,1000),sal2,color='red',label='Segunda derivada estimada',linewidth=4)
plt.plot(np.linspace(-10,10,1000),sal3,color='green',label='Funcion Original',linewidth=4)
plt.legend()
plt.grid((50,50),color='black',alpha=0.5)
plt.title("Diferencias finitas para encontrar primera y segunda derivada de x^4+1")
plt.savefig("Diferencias_finitas.jpg")
```
