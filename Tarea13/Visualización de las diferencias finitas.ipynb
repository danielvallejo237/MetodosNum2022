```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("output.txt","r") as f:
    lines=f.readlines()
arr=np.zeros((len(lines),12),dtype=float)
for i,l in enumerate(lines):
    arr[i]=np.array(l.split(',')[:-1])
```

```python
sz=0.01
fig,ax=plt.subplots(1,3,figsize=(15,8))
ax[0].plot(arr.T[0],color='red',label='GroundTruth')
ax[0].plot(arr.T[3],color='magenta',label='Centrada')
ax[0].plot(arr.T[6],color='green',label='Izquierda')
ax[0].plot(arr.T[9],color='blue',label='Derecha')
ax[0].legend()
ax[0].set_title("Derivada C1 con tamaño de paso: {}".format(sz))

ax[1].plot(arr.T[1],color='red',label='GroundTruth')
ax[1].plot(arr.T[4],color='magenta',label='Centrada')
ax[1].plot(arr.T[7],color='green',label='Izquierda')
ax[1].plot(arr.T[10],color='blue',label='Derecha')
ax[1].legend()
ax[1].set_title("Derivada C2 con tamaño de paso: {}".format(sz))

ax[2].plot(arr.T[2],color='red',label='GroundTruth')
ax[2].plot(arr.T[5],color='magenta',label='Centrada')
ax[2].plot(arr.T[8],color='green',label='Izquierda')
ax[2].plot(arr.T[11],color='blue',label='Derecha')
ax[2].legend()
ax[2].set_title("Derivada C3 con tamaño de paso: {}".format(sz))

fig.suptitle("Comparación del efecto de los diferentes tamaños de paso en cada una de las derivadas",fontsize=20)

fig.savefig("Comparativo_sz"+str(sz)+".jpg")
```
